Bootstrap code

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
from bootstrap import bootstrap
```

Bootstrap test

```python
# Test array
np.random.seed(42)
X = np.random.randint(0, 10, 25)

print(np.sort(X))

# Perform bootstrapping
boots_100 = bootstrap(X, n_samples=25, n_bootstraps=100)
boots_1000 = bootstrap(X, n_samples=25, n_bootstraps=1000)

# Create subplots
fig, axis = plt.subplots(3, 1, figsize=(7.5, 10))
ax_X, ax_boot_100, ax_boot_1000 = axis.flatten()

fontsize = 16

ax_X.hist(X, color="green", edgecolor="black", alpha=0.7)
ax_X.axvline(np.mean(X), color="red", linestyle="--", linewidth=2, label=f"Mean")
ax_X.set_title(f"a) Original dataset (mean={np.mean(X):.3f})", fontsize=fontsize)
ax_X.set_xlabel("Value", fontsize=fontsize)
ax_X.tick_params(axis="both", labelsize=fontsize-2)
ax_X.legend(fontsize=fontsize)

ax_boot_100.hist(np.mean(boots_100, axis=1), edgecolor="black", bins=20, alpha=0.8)
ax_boot_100.axvline(np.mean(boots_100), color='red', linestyle='--', linewidth=2)
ax_boot_100.set_title(f"b) 100 Bootstrap samples (mean={np.mean(boots_100):.3f})", fontsize=fontsize)
ax_boot_100.set_xlabel("Bootstrap Mean", fontsize=fontsize)
ax_boot_100.tick_params(axis='both', labelsize=fontsize-2)

ax_boot_1000.hist(np.mean(boots_1000, axis=1), edgecolor="black", bins=20, alpha=0.8)
ax_boot_1000.axvline(np.mean(boots_1000), color='red', linestyle='--', linewidth=2)
ax_boot_1000.set_title(f"c) 1000 Bootstrap samples (mean={np.mean(boots_1000):.3f})", fontsize=fontsize)
ax_boot_1000.set_xlabel("Bootstrap Mean", fontsize=fontsize)
ax_boot_1000.tick_params(axis='both', labelsize=fontsize-2)

fig.suptitle("Bootstrap Sampling", fontsize=fontsize+2, y=0.995)
plt.tight_layout()
plt.show()
```
